```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load and read the dataset into panda dataframe
#Read statistics
cars_df = pd.read_csv('../input/cars-dataset/2. Cars Data1.csv')
cars_df.head()
```

```python
#Get more information about the dataset
cars_df.info()
```

```python
#Check for missing values
cars_df.isnull().sum()
```

The missing values are few and so can be filled up using mean and/or mode.

```python
#To fill columns with mean
mean = cars_df.filter(['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']).mean()
cols = ['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']
cars_df[cols]=cars_df[cols].fillna(cars_df.mean().iloc)
```

```python
#To fill columns with mode
mode = cars_df.filter(['Make','Model','Type','Origin','DriveTrain','MSRP','Invoice']).mode()
cols = ['Make','Model','Type','Origin','DriveTrain','MSRP','Invoice']
cars_df[cols]=cars_df[cols].fillna(cars_df.mode().iloc[0])
```

```python
cars_df.isnull().sum()
```

```python
#Check for sum of duplicate values
sum(cars_df.duplicated())
```

```python
#Drop duplicates
cars_df.drop_duplicates(inplace=True)
```

```python
cars_df.info()
```

```python
#Convert object to integer
cols = ['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']
cars_df[cols] = cars_df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```
