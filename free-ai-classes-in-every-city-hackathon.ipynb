```python
#Import libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn
import sklearn

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
#Import data

test = pd.read_csv('../input/test-and-train/Tests.csv')
train = pd.read_csv('../input/test-and-train/Trains.csv')

display(train.head())
display(test.head())
```

Clean the dataset

```python
#Check for columns for train
train.columns
```

```python
#Check for columns for test
test.columns
```

```python
#Check for missing values for train
train.isnull().sum()
```

```python
# fill up missing values in Distance with the mode

train['Distance'] = train['Distance'].str.replace(',','')

mode_value = train['Distance'].mode()[0]
print(mode_value)

train['Distance'].fillna(mode_value, inplace=True)

train['Distance'] = train['Distance'].astype(int)

train.head()
```

```python
# fill up missing values in Year with the mode

train['Year'] = train['Year'].str.replace(',','')

mode_value = train['Year'].mode()[0]
print(mode_value)

train['Year'].fillna(mode_value, inplace=True)

train['Year'] = train['Year'].astype(int)

train.head()
```

```python
# fill up missing values in Type with the mode
mode_value = train['Type'].mode()[0]
print(mode_value)

train['Type'].fillna(mode_value, inplace=True)
```

```python
# fill up missing values in Amount (Million Naira) with the mean
mean_value = train['Amount (Million Naira)'].mean()
print(mean_value)

train['Amount (Million Naira)'].fillna(mean_value, inplace=True)
```

```python
#Check again for missing values for train

display(train.isnull().sum())

display(train.head())
```

```python
#Check for missing values for test

test.isnull().sum()
```

```python
# fill up missing values in Distance with the mean
mean_value = test['Distance'].mean()
print(mean_value)

test['Distance'].fillna(mean_value, inplace=True)
```

```python
# fill up missing values in Type with the mode
mode_value = test['Type'].mode()[0]
print(mode_value)

test['Type'].fillna(mode_value, inplace=True)
```

```python
test['Year'] = test['Year'].str.replace(',','')

mode_value = test['Year'].mode()[0]
print(mode_value)

test['Year'].fillna(mode_value, inplace=True)

test['Year'] = test['Year'].astype(int)

test.head()
```

```python
#Check for missing values again 

display(test.isnull().sum())

display(test.head())
```

```python
display(train.head())
display(test.head())
```

```python
#Label Encoding

featured = [ 'VehicleID','Location','Maker','Model','Colour','Type' ]

for feature in featured:
  train[f"{feature}_cat"] = train[feature].astype('category')
  train[f"{feature}_cat"] = train[f"{feature}_cat"].cat.codes


# Read more on Pandas get_dummies

display(train.head())

#for test
featured = [ 'VehicleID','Location','Maker','Model','Colour','Type' ]

for feature in featured:
  test[f"{feature}_cat"] = test[feature].astype('category')
  test[f"{feature}_cat"] = test[f"{feature}_cat"].cat.codes



display(test.head())
```

```python
# Drop the reductant features since Label encoding have been done
train.drop(['VehicleID','Location','Maker','Model','Colour','Type'], axis=1, inplace=True)
display(train.head())

#for test
test.drop(['VehicleID','Location','Maker','Model','Colour','Type'], axis=1, inplace=True)
display(test.head())
```

Data Segmentation

```python
y = train['Amount (Million Naira)'] # Target
X = train.drop('Amount (Million Naira)', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

Train Model

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

Make Predictions

```python
reg.predict(X_test)
```

Evaluate Model

```python
from sklearn.metrics import mean_squared_error

# we are using mean_absolute_error because 
# this is a regression model

y_pred1 = reg.predict(X_test)

print("MSE",mean_squared_error(y_test,y_pred1))
```

```python
mse=mean_squared_error(y_test,y_pred1)
rmse=mean_squared_error(y_test,y_pred1,squared=False)
rmse
```
